```python
df_1 =  pd.read_excel('KPMG.xlsx', 'NewData')
```

```python
df_1.info()
```

```python
R_score = [] 
for value in df_1["recency"]: 
    if value > 88: 
        R_score.append("1") 
    elif value <= 88 and value > 45: 
        R_score.append("2")
    elif value <= 45 and value > 18: 
        R_score.append("3") 
    else: 
        R_score.append("4") 
       
df_1["R_score"] = R_score    
print(df_1) 
```

```python
F_score = [] 
for value in df_1["product_id"]: 
    if value > 7: 
        F_score.append("1") 
    elif value <= 7 and value > 5: 
        F_score.append("2")
    elif value <= 5 and value > 4: 
        F_score.append("3") 
    else: 
        F_score.append("4") 
       
df_1["F_score"] = F_score    
print(df_1) 
```

```python
M_score = [] 
for value in df_1["profit"]: 
    if value > 4123.1: 
        M_score.append("1") 
    elif value <= 4123.1 and value > 2799.26: 
        M_score.append("2")
    elif value <= 2799.26 and value > 1797.95: 
        M_score.append("3") 
    else: 
        M_score.append("4") 
       
df_1["M_score"] = M_score    
print(df_1) 
```

```python
df_1.head(10)
```

```python
#a = df_1['R_score'].apply(lambda x : x*100)
#a
```

```python
df_1["R_score"] = df_1["R_score"].astype(str).astype(int)
```

```python
df_1["F_score"] = df_1["F_score"].astype(str).astype(int)
```

```python
df_1["M_score"] = df_1["M_score"].astype(str).astype(int)
```

```python
df_1['RFM value'] = ((df_1['R_score']*100) + (10*df_1['F_score']) + df_1['M_score'])
```

```python
df_1['RFM value'].describe()
```

```python
Cust_title = [] 
for value in df_1["RFM value"]: 
    if value > 411: 
        Cust_title.append("Platinum") 
    elif value <= 411 and value > 311: 
        Cust_title.append("Gold")
    elif value <= 311 and value > 211: 
        Cust_title.append("Silver") 
    else: 
        Cust_title.append("Bronze") 
       
df_1["Customer_Title"] = Cust_title    
print(df_1) 
```
